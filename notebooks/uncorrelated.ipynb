```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.DataFrame()
```

```python
df['one'] = np.random.normal(0, 1, 1000000)
df['two'] = np.random.normal(0, 1, 1000000)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df.corr()
```

```python
plt.scatter(df['one'], df['two'], c = 'y', edgecolor = 'g')
plt.xlabel('one')
plt.ylabel('two')
plt.axes().set_aspect('equal')
```
