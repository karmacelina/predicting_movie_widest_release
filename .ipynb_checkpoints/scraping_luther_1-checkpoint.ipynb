```python
def get_wiki_movies(soupObjects, start = 1980, end = 2013):
    
    title = pd.Series([], name = 'title')
    director = pd.Series([], name = 'director')
    cast = pd.Series([], name = 'cast')
    genre = pd.Series([], name = 'genre')
    studios = pd.Series([], name = 'studios')
    release_date = pd.Series([], name = 'release date')
    release_year = pd.Series([], name = 'release year')
    
    years = [str(i) for i in range(start, end + 1)]
    
    for year in years:
        
        for table in soupObjects[year].find_all(class_ = 'wikitable'):
            
            for row in table.find_all('tr'):
                
                r = row.find_all('td')
                if not r:
                    continue
                
                try:
                    title = title.append(pd.Series(r[0].text))
                except IndexError:
                    title = title.append(pd.Series('NA'))
                
                try:
                    director = director.append(pd.Series(r[1].text))
                except IndexError:
                    director = director.append(pd.Series('NA'))
                
                try:
                    cast = cast.append(pd.Series(r[2].text))
                except IndexError:
                    cast = cast.append(pd.Series('NA'))
                
                try:
                    genre = genre.append(pd.Series(r[3].text))
                except IndexError:
                    genre = genre.append(pd.Series('NA'))
                
                try:
                    studios = studios.append(pd.Series(r[4].text))
                except IndexError:
                    studios = studios.append(pd.Series('NA'))
                
                
                try:
                    release_date = release_date.append(pd.Series(r[5].text))
                except IndexError:
                    release_date = release_date.append(pd.Series('NA'))
                    
                release_year = release_year.append(pd.Series(year))
                #
                
    # concatenate series into dataframe            
    df = pd.concat([title, director, cast, genre, studios, release_date, release_year], axis=1)
    #df = pd.concat([title, release_year], axis=1)
    df.columns = ['title','director','cast','genre','studios','release date','release year']
    #df.columns = ['title','release year']
    df.index = df['title']
  
    return df
```

```python
def get_movie_value(soup, field_name):
    '''Grab a value from boxofficemojo HTML
    
    Takes a string attribute of a movie on the page and
    returns the string in the next sibling object
    (the value for that attribute)
    or None if nothing is found.
    '''
    obj = soup.find(text=re.compile(field_name))
    if not obj: 
        return None
    # this works for most of the values
    next_sibling = obj.findNextSibling()
    if next_sibling:
        return next_sibling.text # RETURN TEXT OF NEXT SIBLING
    else:
        return None
```

```python
def get_movie_value_next(soup, field_name):
    '''Grab a value from boxofficemojo HTML
    
    Takes a string attribute of a movie on the page and
    returns the string in the next sibling object
    (the value for that attribute)
    or None if nothing is found.
    '''
    obj = soup.find(text=re.compile(field_name))
    
    if obj:
        return obj.next.text
    else:
        return None
    
```

```python
def get_movie_value_next_next(soup, field_name):
    '''Grab a value from boxofficemojo HTML
    
    Takes a string attribute of a movie on the page and
    returns the string in the next sibling object
    (the value for that attribute)
    or None if nothing is found.
    '''
    obj = soup.find(text=re.compile(field_name))
    
    if obj:
        return obj.next.next.text
    else:
        return None
```

```python
import dateutil.parser

def to_date(datestring):
    date = dateutil.parser.parse(datestring)
    return date

def money_to_int(moneystring):
    moneystring = moneystring.replace('$', '').replace(',', '')
    return int(moneystring)

def money_to_int_2(moneystring):
    if moneystring != 'N/A':
        moneystring = moneystring.replace('$', '').replace('.','').replace(',','').replace(' million', '000000')
        return int(moneystring)
    else:
        return None
    
def strip_xa0(s):
    """ Removes \xa0 unicode character from string """
    s = s.text
    s_no_xa0 = s.replace(u'\xa0', '') 
    return s_no_xa0

def strip_and_return_int(s):
    t = s.replace('$', '').replace(',','')
    t = int(t)
    return t

#def list_actors('Actor'):
#    actor = get_movie_value_next(soup,string)
#    v = re.sub('([a-z()])([A-Z])', '\g<1>,\g<2>', actor)
#    actors = re.sub('[(*\']','', v)
#    actors = actors.split(",
#    return actors
#

#def list_producers('Producer'):
#    t = get_movie_value_next(soup, string)
#    list_producers = re.sub('([a-z()])([A-Z])', '\g<1>, \g<2>', t).split(",") 
#    return list_producers

def runtime_to_minutes(runtimestring):
    runtime = runtimestring.split()
    try:
        minutes = int(runtime[0])*60 + int(runtime[2])
        return minutes
    except:
        return None

#def get_title(string = 'title'):
#    title_string = soup.find(string).text
#    title = title_string.split('(')[0].strip()
#    return title
#
#def list_names(string):
#    t = get_movie_value_next(soup, string)
#    list_names = re.sub('([a-z()])([A-Z])', '\g<1>, \g<2>', t).split(",") 
#    return list_names
```

```python
def data_from_movie(soup): # a function of what? soup? 
    
    # title
    title_string = soup.find('title').text
    title = title_string.split('(')[0].strip()
    
    # release date 
    try:
        raw_release_date = get_movie_value(soup,'Release Date')
        release_date = to_date(raw_release_date)
    except AttributeError:
        release_date = float('NaN')
    
    # domestic gross
    try:
        raw_domestic_total_gross = get_movie_value(soup,'Domestic Total')
        domestic_total_gross = money_to_int(raw_domestic_total_gross)
    except AttributeError:
        domestic_total_gross = float('NaN')
    
    # opening weekend
    try:
        opening_weekend_gross = get_movie_value_next(soup,'Opening Weekend')
        opening_weekend_gross = money_to_int(opening_weekend_gross)
    except AttributeError:
        opening_weekend_gross = float('NaN')
    
    #production budget
    try:
        production_budget = get_movie_value(soup, 'Production Budget')
        production_budget = money_to_int_2(production_budget)
    except AttributeError:
        production_budget = float('NaN')
    
    #movie runtime
    try:
        raw_runtime = get_movie_value(soup,'Runtime')
        runtime = runtime_to_minutes(raw_runtime)
    except AttributeError:
        runtime = float('NaN')
        
    # max number of theaters movie was played in
    try:
        widest_release = get_movie_value_next_next(soup, 'Widest Release')
        widest_release = strip_and_return_int(widest_release.split()[0])
    except AttributeError:
        widest_release = float('NaN')
      
    
    # actors
    try:
        actor = get_movie_value_next(soup,'Actor')
        v = re.sub('([a-z()])([A-Z])', '\g<1>,\g<2>', actor)
        actors = re.sub('[(*\']','', v)
        actors = actors.split(",") 
    except TypeError:
        actors = float('NaN')
    
    # director
    try:
        director = get_movie_value_next(soup,'Director')
        director = re.sub('([a-z()])([A-Z])', '\g<1>, \g<2>', director).split(",") 
    except TypeError:
        director = float('NaN')

    # producer 
    try:
        producer = get_movie_value_next(soup, 'Producer')
        producers = re.sub('([a-z()])([A-Z])', '\g<1>, \g<2>', producer).split(",") 
    except TypeError:
        producers = float('NaN')
    except AttributeError:
        producers = float('NaN')
    
    # rating
    rating = get_movie_value(soup,'MPAA Rating')
    
    # genre
    genre = get_movie_value(soup, 'Genre[^a-z]')  
    
    
    
    headers = ['movie title', 'domestic total gross',
           'release date', 'runtime (mins)', 'rating', 
           'opening weekend gross', 'director', 'production budget', 
           'widest release', 'actors', 'producers', 'genre']
    
    movie_dict = dict(zip(headers, [title, domestic_total_gross,
                                    release_date,runtime,rating, 
                                    opening_weekend_gross, director, 
                                    production_budget, widest_release, 
                                    actors, producers, genre]))
    
    return movie_dict
```
